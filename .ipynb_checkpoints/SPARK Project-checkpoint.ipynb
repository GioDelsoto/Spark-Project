# Import libraries

```python
from pyspark.sql import SparkSession
from pyspark.ml import Pipeline
from pyspark.ml.pipeline import PipelineModel
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.feature import StandardScaler

from pyspark.sql.functions import col, count
from pyspark.ml.evaluation import RegressionEvaluator

from functools import reduce

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Step 1 - Preparing the data

```python
# Create a spark session
spark = SparkSession.builder.appName('Spark project').master("local").getOrCreate()
```

```python
#Download the file

import requests

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMSkillsNetwork-BD0231EN-Coursera/datasets/NASA_airfoil_noise_raw.csv"
response = requests.get(url)

# Salvando o arquivo localmente
with open("NASA_airfoil_noise_raw.csv", "wb") as file:
    file.write(response.content)

print("Download completed")
```

```python
df = spark.read.csv('NASA_airfoil_noise_raw.csv', header = True, inferSchema = True)
df.show(5)
```

```python
df.printSchema()
```

```python
# Number of rows
num_rows = df.count()

# Number of columns
num_cols = len(df.columns)

print(f"Shape: ({num_rows}, {num_cols})")
```

```python
df = df.withColumnRenamed('SoundLevel','SoundLevelDecibels')
```

## 1.2 - Duplicated values

```python
duplicates = (
    df.groupBy(df.columns)
    .agg(count("*").alias("count"))
    .filter(col("count") > 1)
)

# Show duplicate rows
duplicates.show()
```

```python
df = df.dropDuplicates()
```

```python
num_rows = df.count()
print(f"Number of rows: {num_rows}")
```

## 1.3 Null values

```python
rows_with_nulls = df.filter(
    reduce(lambda x, y: x | y, (col(c).isNull() for c in df.columns))
)

# Show rows with null values
rows_with_nulls.show()
```

```python
#We only have 4 rows with empty values. We can just drop it

df=df.dropna()
```

## 1.4 Dataframe statistics

```python
df.describe().show()
```

## 1.5 Save the data

```python
df.write.mode("overwrite").parquet("NASA_airfoil_noise_cleaned.parquet")
```

# Step 3 - Build the Model

```python
df = spark.read.parquet("NASA_airfoil_noise_cleaned.parquet")
```

```python
df.count()
```

## 3.1 - Create pipeline

```python
# Assembler
feature_cols = ['Frequency','AngleOfAttack','ChordLength','FreeStreamVelocity','SuctionSideDisplacement']

assembler = VectorAssembler(inputCols = feature_cols, outputCol = 'features')
```

```python
#Scaler
scaler = StandardScaler(inputCol = 'features', outputCol = 'scaledFeatures')
```

```python
lr = LinearRegression(featuresCol='features', labelCol='SoundLevelDecibels')
```

```python
pipeline = Pipeline(stages = [assembler,scaler,lr])
```

## 3.2 Split the data

```python
train_df, test_df = df.randomSplit([0.7, 0.3], seed = 42)
```

## 3.3 Fit the data

```python
model = pipeline.fit(train_df)
```

```python
ps = [str(x).split("_")[0] for x in pipeline.getStages()]

print("Pipeline Stage 1 = ", ps[0])
print("Pipeline Stage 2 = ", ps[1])
print("Pipeline Stage 3 = ", ps[2])

print("Label column = ", lr.getLabelCol())
```

# 4 Model evaluation

```python
predictions = model.transform(test_df)
```

```python
evaluator = RegressionEvaluator(predictionCol="prediction", labelCol="SoundLevelDecibels", metricName="mse")
mse = evaluator.evaluate(predictions)
print(f"MSE: {mse:.4f}")
```

```python
evaluator = RegressionEvaluator(predictionCol="prediction", labelCol="SoundLevelDecibels", metricName="mae")
mae = evaluator.evaluate(predictions)
print(f"MAE: {mae:.4f}")
```

```python
evaluator = RegressionEvaluator(predictionCol="prediction", labelCol="SoundLevelDecibels", metricName="r2")
r2 = evaluator.evaluate(predictions)
print(f"R-square: {r2:.4f}")
```

## Intercept and coefficients

```python
intercept = model.stages[-1].intercept
print(f"Intercept: {intercept}")

# Get the coefficients
coefficients = model.stages[-1].coefficients

print(f"Coefficients: {list(zip(feature_cols,coefficients))}")
```

### Visualize the results

```python
predictions_df = predictions.select("SoundLevelDecibels", "prediction")
predictions_pd = predictions_df.toPandas()
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(predictions_pd['SoundLevelDecibels'], predictions_pd['prediction'], alpha=0.6)

# Add title and axis labels
plt.title('Actual vs. Predicted Values')
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')

plt.plot([predictions_pd['SoundLevelDecibels'].min(), predictions_pd['SoundLevelDecibels'].max()],
         [predictions_pd['SoundLevelDecibels'].min(), predictions_pd['SoundLevelDecibels'].max()],
         color='red', linestyle='--')

plt.show()
```
